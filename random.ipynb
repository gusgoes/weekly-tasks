```python
#Numerical arrays
import numpy as np

#Plot

import matplotlib.pylab as plt
```

***
### Generate Random
***

```python
rng = np.random.default_rng()
# Generate one random float uniformly distributed over the range [0, 1)

```

```python
rng.random()  
```

```python
rng.standard_normal(10) 
```

```python
# Generate an array of 5 integers uniformly over the range [0, 10).
rng.integers(low=0, high=10, size=5)
```

```python
# Generate 10.0000 integers

x = rng.integers(low=1, high=10, size=10000, endpoint=True)
x
```

```python
# Get the counts
values, counts = np.unique(x, return_counts=True)
values, counts
```

```python
#Bar Chart
plt.bar(values, counts)
```

***
### Uniform Distribution
***

```python
y = rng.uniform(low=0.0, high=10.0, size=100)
y
```

```python
values, counts = np.unique(y, return_counts=True)

values, counts, values.size, counts.size
```

```python
#Bar Chart
plt.bar(values, counts, edgecolor='black')
```

```python
# Round values i x

x_rounded = np.round(y, decimals=0)

values, counts = np.unique(x_rounded, return_counts=True)

values, counts, values.size, counts.size
```

```python
plt.bar(values, counts, edgecolor='black')
```
